```python
def load_series_dict(series_dict_path):
    with open(series_dict_path, 'r') as f:
        json_dict = json.load(f)
        # convert dict keys, values to int
        json_dict = {int(k): int(v) for k, v in json_dict.items()}
        return json_dict
```

```python
accumulated_dictionary = accumulated_series_dicti(series_dict_path)
```

```python
for i in accumulated_dictionary.keys():
    accumulated_dictionary[i] = accumulated_dictionary[i]*2
```

```python
onlyconnect_json = {
    "season_to_walls_map": json.load(open(series_dict_path, 'r')),
    "dataset": []
}
```

```python
for i in onlyconnect_json['season_to_walls_map'].keys():
    onlyconnect_json['season_to_walls_map'][i] = onlyconnect_json['season_to_walls_map'][i]*2
```

```python
for season in range(1, 15 + 1):
    for idx in range(0, accumulated_dictionary[season]):
        words = onlyconnect[int((season * 16) + (idx * 16) - 16):int((season * 16) + (idx * 16))]['Names'].tolist()
        connections = onlyconnect[int((season * 16) + (idx * 16) - 16):int((season * 16) + (idx * 16))]['Answer'].tolist()
        connections_unique = list(dict.fromkeys(connections))
        human_wall = onlyconnect[int((season * 16) + (idx * 16) - 16):int((season * 16) + (idx * 16))]['Human wall'].tolist()
        human_connection = onlyconnect[int((season * 16) + (idx * 16) - 16):int((season * 16) + (idx * 16))]['Human Connection'].tolist()
        onlyconnect_json['dataset'].append({
            #generate a unique id for each wall
            "wall_id": 0,
            "season": season,
            "episode": idx+1,
            "words": words,
            "gt_connections": connections_unique,
            "groups": {
                "group_1": {
                    "group_id": 0,
                    "gt_words": words[:4],
                    "gt_connection": connections[0],
                    "human_performance": {
                        "grouping": human_wall[0],
                        "connection": human_connection[0]
                    }
                },
            "group_2": {
                    "group_id": 0,
                    "gt_words": words[4:8],
                    "gt_connection": connections[4],
                    "human_performance": {
                        "grouping": human_wall[4],
                        "connection": human_connection[4]
                    }
                },
            "group_3": {
                    "group_id": 0,
                    "gt_words": words[8:12],
                    "gt_connection": connections[8],
                    "human_performance": {
                        "grouping": human_wall[8],
                        "connection": human_connection[8]
                    }
                },
        "group_4": {
                    "group_id": 0,
                    "gt_words": words[12:],
                    "gt_connection": connections[12],
                    "human_performance": {
                        "grouping": human_wall[12],
                        "connection": human_connection[12]
                    }
                }
            },
            "overall_human_performance": {
                "grouping": [human_wall[0], human_wall[4], human_wall[8], human_wall[12]],
                "connections": [human_connection[0], human_connection[4], human_connection[8], human_connection[12]]
            }
    })
```
